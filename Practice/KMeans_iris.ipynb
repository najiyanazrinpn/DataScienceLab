```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data[:, :2]
Y = iris.target
```

```python
import matplotlib.pyplot as plt

x_axis = X[:, 0] #ALL sepal length values in the data
y_axis = X[:, 1] #All sepal width values in the data

plt.figure(figsize=(6,6))
plt.scatter(x_axis, y_axis, c=Y, cmap='prism')
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Iris Dataset", fontsize=20)
plt.show()
print()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3, random_state=21)
km.fit(X)
centers = km.cluster_centers_
new_labels = km.labels_
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(x_axis, y_axis, c=Y, cmap='prism')
axes[0].set_title("Actual")
axes[0].set_xlabel("Sepal Length")
axes[0].set_xlabel("Sepal Width")

axes[1].scatter(x_axis, y_axis, c=new_labels, cmap='prism')
axes[1].set_title("Predicted")
axes[1].set_xlabel("Sepal Length")
axes[1].set_xlabel("Sepal Width")

plt.show()
```
