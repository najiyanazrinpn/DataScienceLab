```python
import pandas as pd
data = pd.read_csv("dataset.csv")
print(data)
```

```python
X = data[['gives_birth', 'aquatic_animal', 'aerial_animal', 'has_legs']]
Y = data[['class_label']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=1)
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(X_train, Y_train)
y_pred = clf.predict(X_test)
print(y_pred)
new_case = [[1,1,0,0]]
y_new = clf.predict(new_case)
print(y_new)
```

```python
from sklearn import metrics
print("Accuracy", metrics.accuracy_score(Y_test, y_pred))
```

```python
feature_names = ['gives_birth', 'aquatic_animal', 'aerial_animal', 'has_legs']
target_names = ['mammal', 'reptile', 'fish', 'bird', 'amphibian']
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,12))
tree.plot_tree(clf, class_names = target_names, feature_names = feature_names)
plt.show()
```
