```python
import pandas as pd
dataset = pd.read_csv("dataset")
X = dataset[['GivesBirth', 'AquaticAnimal']]
Y = dataset[['ClassLabel']]
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 5)
km.fit(X)
centers = km.cluster_centers_
new_labels = km.labels_
```

```python
import matplotlib.pyplot as plt
x_axis = X['GivesBirth']
y_axis = X['AquaticAnimal']
fig, axes = plt.subplots(1,2,figsize=(12,6))
axes[0].scatter(x_axis, y_axis, c=Y.values.ravel(), cmap='prism')
axes[1].scatter(x_axis, y_axis, c=new_labels, cmap='prism')
```
