```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,18])
y1 = np.array([3,10,12,20])
y2 = np.array([10,20,40,12])
plt.plot(x,y1,label="India")
plt.plot(x,y2,label="sK")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,6,18])
y1 = np.array([3,10,12,20])
y2 = np.array([10,20,40,12])

plt.subplot(1,2,1)
plt.plot(x,y1,label="India")
plt.legend()

plt.subplot(1,2,2)
plt.plot(x,y2,label="SK")
plt.legend()

plt.show()
```

```python
x_pts = ["java", "Python", "PHP", "JavaScript", "C#", "C++"]
y_pts = [22.2, 17.6, 8.8, 8, 77, 6.7]
color = ['red','yellow','green','cyan','purple','gray']
plt.bar(x_pts, y_pts, color=color)
# plt.barh(x_pts, y_pts)
```

```python
y1 = np.array([22, 30, 35, 35, 26])
y2 = np.array([25, 32, 30, 35, 29])
grp = np.array(['G1', 'G2', 'G3', 'G4', 'G5'])
x = np.arange(len(grp))
plt.bar(x-.125, y1, width=.25, label="Men")
plt.bar(x+.125, y2, width=.25, label="Women")
plt.legend()
plt.xticks(x,grp)
plt.show()
```
