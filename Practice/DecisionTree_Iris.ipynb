```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
Y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
d = DecisionTreeClassifier()
d.fit(X_train, Y_train)

import inspect
print(inspect.signature(DecisionTreeClassifier))
```

```python
t = DecisionTreeClassifier(criterion='entropy')
t.fit(X_train, Y_train)
```

```python
sample = [[4,3,3,2]]
Y_pred1 = d.predict(sample)
print(iris.target_names[Y_pred1])
print(iris.feature_names)
```

```python
from sklearn import metrics
Y_pred = d.predict(X_test)
print(metrics.accuracy_score(Y_test, Y_pred))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 20))

tree.plot_tree(t, feature_names = iris.feature_names, class_names = iris.target_names, filled=True, rounded=True, fontsize=15)
plt.show()
```
