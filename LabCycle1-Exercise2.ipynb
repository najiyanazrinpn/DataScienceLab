```python
'''Write Python program to create two matrices (read values from user) and find the following
1. Dot Product
2. Transpose
3. Trace
4. Rank
5. Determinant
6. Inverse
7. Eigen values and eigen vectors'''

import numpy as np
n1 = int(input("Enter the size of the matrix1: "))
print("Enter "+str(n1*n1)+" elements: ")

a1 = [[0] * n1 for i in range(n1)]

for i in range(n1):
    for j in range(n1):
        a1[i][j] = int(input())
        
n2 = int(input("Enter the size of the matrix1: "))
print("Enter "+str(n2*n2)+" elements: ")

a2 = [[0] * n2 for i in range(n2)]

for i in range(n2):
    for j in range(n2):
        a2[i][j] = int(input())

x = np.matrix(a1)
y = np.matrix(a2)

print("Matrix1:\n",x)
print("Matrix2:\n",y)
if(n1 == n2):
    print("Matrix 1 * Matrix 2:\n", np.dot(x,y))
else:
    print("Invalid size")

print("Transpose of matrix1:\n",np.transpose(x))
print("Transpose of matrix2:\n",np.transpose(y))

print("Trace of matrix1:\n",x.trace())
print("Trace of matrix2:\n",y.trace())

print("Rank of matrix1:\n",np.linalg.matrix_rank(x))
print("Rank of matrix2:\n",np.linalg.matrix_rank(y))

print("Determinant of matrix1:\n",np.linalg.det(x))
print("Determinant of matrix2:\n",np.linalg.det(y))

if((np.linalg.det(x)//1)!=0):
    print("Inverse of matrix1:\n",np.linalg.inv(x))
else:
    print("Inverse of matrix1 doesn't exist")

if((np.linalg.det(y)//1)!=0):
    print("Inverse of matrix2:\n",np.linalg.inv(y))
else:
    print("Inverse of matrix2 doesn't exist")


print("Eigenvalues and eigenvectors of matrix1:\n",np.linalg.eig(x))
print("Eigenvalues and eigenvectors of matrix2:\n",np.linalg.eig(y))
```
